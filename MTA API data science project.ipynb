```python
#check feed
feed.ParseFromString(response.content)
print(feed.entity)
```

```python
dict_obj = MessageToDict(feed)
print('Dict:')
dict_obj
```

```python
json_obj = MessageToJson(feed)
print('Json:')
json_obj
```

```python
from pandas.io.json import json_normalize
j=json.loads(json_obj)
#data = pd.DataFrame.from_dict(j, orient='index')
df1 = json_normalize(j['entity'])
print(df1)
```

```python
dfclean = df1.rename(columns={'alert.headerText.translation':'alert',
                    'tripUpdate.stopTimeUpdate':'tripStopTime','tripUpdate.trip.routeId':'tripRouteId','tripUpdate.trip.startDate':'tripStartDate','tripUpdate.trip.tripId':'tripId',
                    'vehicle.currentStatus':'curStatus','vehicle.currentStopSequence':'curStopSeq','vehicle.stopId':'stopId','vehicle.timestamp':'timestamp','vehicle.trip.routeId':'routeId','vehicle.trip.startDate':'startDate','vehicle.trip.tripId':'tripId'})
```
